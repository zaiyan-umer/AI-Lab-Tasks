```python
# --- Task 1 ---
import random

class Component:
    def __init__(self, name):
        self.name = name
        self.safe = random.choice([True, False])

    def is_safe(self):
        return self.safe

    def patch(self):
        if not self.safe:
            self.safe = True
            print(f"{self.name} patched")

class SecuritySystem:
    def __init__(self, component_names):
        self.components = [Component(name) for name in component_names]

    def display_state(self, title):
        print(f"\n{title}")
        for comp in self.components:
            print(comp.name, "SAFE" if comp.is_safe() else "VULNERABLE")

    def scan_system(self):
        print("\nScanning system:")
        vulnerable = []
        for comp in self.components:
            if comp.is_safe():
                print(comp.name, "is secure")
            else:
                print(comp.name, "is vulnerable")
                vulnerable.append(comp)
        return vulnerable

    def patch_vulnerabilities(self, vulnerable_components):
        print("\nPatching vulnerabilities:")
        for comp in vulnerable_components:
            comp.patch()
            
# main component (test)
components = ['A','B','C','D','E','F','G','H','I']
system = SecuritySystem(components)

system.display_state("Initial system state:")
vulnerable = system.scan_system()
system.patch_vulnerabilities(vulnerable)
system.display_state("Final system state:")
```

```python
# --- Task 2 ---
import random

class Server:
    def __init__(self, name):
        self.name = name
        self.state = random.choice(['Underloaded', 'Balanced', 'Overloaded'])

    def is_overloaded(self):
        return self.state == 'Overloaded'

    def is_underloaded(self):
        return self.state == 'Underloaded'

    def balance(self):
        self.state = 'Balanced'

class LoadBalancer:
    def __init__(self, server_names):
        self.servers = [Server(name) for name in server_names]

    def display_load(self, title):
        print(f"\n{title}")
        for server in self.servers:
            print(server.name, server.state)

    def balance_load(self):
        overloaded = []
        underloaded = []

        for server in self.servers:
            if server.is_overloaded():
                overloaded.append(server)
            elif server.is_underloaded():
                underloaded.append(server)

        for o in overloaded:
            if underloaded:
                u = underloaded.pop(0)
                o.balance()
                u.balance()
                
# main component (test)
servers = ['S1', 'S2', 'S3', 'S4', 'S5']
lb = LoadBalancer(servers)

lb.display_load("Initial server load:")
lb.balance_load()
lb.display_load("After load balancing:")
```

```python
# --- Task 3 ---
import random

class BackupTask:
    def __init__(self, name):
        self.name = name
        self.status = random.choice(['Completed', 'Failed'])

    def retry(self):
        if self.status == 'Failed':
            self.status = 'Completed'
            print(f"{self.name} retried and completed")

class BackupSystem:
    def __init__(self, tasks):
        self.tasks = [BackupTask(t) for t in tasks]

    def display_status(self, title):
        print(f"\n{title}")
        for task in self.tasks:
            print(task.name, task.status)

    def retry_failed_backups(self):
        for task in self.tasks:
            task.retry()

# main component (test)
tasks = ['Backup1', 'Backup2', 'Backup3', 'Backup4', 'Backup5']
system = BackupSystem(tasks)

system.display_status("Initial backup status:")
system.retry_failed_backups()
system.display_status("After retrying failed backups:")
```

```python
# --- Task 4 ---
import random

class Component:
    def __init__(self, name):
        self.name = name
        self.state = random.choice(['Safe', 'Low Risk Vulnerable', 'High Risk Vulnerable'])

    def is_safe(self):
        return self.state == 'Safe'

    def patch(self):
        if self.state == 'Low Risk Vulnerable':
            self.state = 'Safe'
            print(f"{self.name} low risk vulnerability patched")
        elif self.state == 'High Risk Vulnerable':
            print(f"{self.name} requires premium service to patch")

class SecuritySystem:
    def __init__(self, component_names):
        self.components = [Component(name) for name in component_names]

    def display_state(self, title):
        print(f"\n{title}")
        for comp in self.components:
            print(comp.name, comp.state)

    def scan_system(self):
        print("\nScanning system:")
        for comp in self.components:
            if comp.is_safe():
                print(comp.name, "is safe")
            else:
                print(comp.name, "has", comp.state)

    def patch_vulnerabilities(self):
        print("\nPatching phase:")
        for comp in self.components:
            comp.patch()
            
# main component (test)
components = ['A','B','C','D','E','F','G','H','I']
system = SecuritySystem(components)

system.display_state("Initial system state:")
system.scan_system()
system.patch_vulnerabilities()
system.display_state("Final system state:")
```

```python
# --- Task 5 ---
import random
import time

class Hospital:
    def __init__(self):
        self.rooms = ['Room101', 'Room102', 'Room103', 'Room104']
        self.medicine_storage = ['MedA', 'MedB', 'MedC', 'MedD']
        self.nurse_station = 'NurseStation'
        self.patient_schedule = {
            'Room101': 'MedA',
            'Room102': 'MedB',
            'Room103': 'MedC',
            'Room104': 'MedD'
        }
        self.staff_available = True

class DeliveryRobot:
    def __init__(self, hospital):
        self.hospital = hospital

    def move_to(self, location):
        print(f"Moving to {location}...")
        time.sleep(0.3)

    def pick_up_medicine(self, med):
        print(f"Picking up {med} from storage...")
        time.sleep(0.3)

    def scan_patient_id(self, room):
        print(f"Scanning patient ID in {room}...")
        time.sleep(0.3)
        # Randomly simulate ID verification failure (10% chance)
        return random.random() > 0.1

    def deliver_medicine(self, room, med):
        print(f"Delivering {med} to {room}...")
        time.sleep(0.3)

    def alert_staff(self, message):
        print(f"Alerting staff: {message}")

    def perform_deliveries(self):
        for room in self.hospital.rooms:
            med = self.hospital.patient_schedule[room]
            self.move_to('MedicineStorage')
            self.pick_up_medicine(med)

            self.move_to(room)
            if self.scan_patient_id(room):
                self.deliver_medicine(room, med)
            else:
                self.alert_staff(f"Patient ID mismatch in {room}")

            if not self.hospital.staff_available:
                self.alert_staff(f"Staff assistance needed in {room}")

# main component (test)
hospital = Hospital()
robot = DeliveryRobot(hospital)

print("Starting hospital delivery robot simulation...\n")
robot.perform_deliveries()
print("\nAll scheduled deliveries completed.")
```

```python
# --- Task 6 ---
class Room:
    def __init__(self, name, status):
        self.name = name
        self.status = status  # "safe" or "fire"

    def has_fire(self):
        return self.status == "fire"

    def extinguish_fire(self):
        if self.has_fire():
            print(f"Fire detected in room {self.name}! Extinguishing...")
            self.status = "safe"
        else:
            print(f"Room {self.name} is safe.")


class FirefightingRobot:
    def __init__(self, building, path):
        self.building = building
        self.path = path
        self.position = None

    def move_and_extinguish(self):
        for room_name in self.path:
            self.position = room_name
            room = self.building[room_name]
            print(f"\nMoving to room {room_name}")
            room.extinguish_fire()
            self.display_building()

    def display_building(self):
        print("\nCurrent building status:")
        grid_order = ['a','b','c','d','e','f','g','h','j']
        for i, name in enumerate(grid_order):
            symbol = "fire" if self.building[name].has_fire() else "safe"
            print(symbol, end="  ")
            if (i + 1) % 3 == 0:
                print()


# Initialize building using normal loop
initial_status = {
    'a': 'safe', 'b': 'safe', 'c': 'fire',
    'd': 'safe', 'e': 'fire', 'f': 'safe',
    'g': 'safe', 'h': 'safe', 'j': 'fire'
}

building = {}
for name in initial_status:
    building[name] = Room(name, initial_status[name])

path = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'j']

# Run simulation
robot = FirefightingRobot(building, path)
robot.move_and_extinguish()

print("\nFinal building status:")
robot.display_building()
```
