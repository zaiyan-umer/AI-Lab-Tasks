# Q1

```python
import pandas as pd

df = pd.read_csv('Mall_Customers.csv')

df
```

```python
X = df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]

X
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4)

kmeans.fit(X)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X.values[:, 0], X.values[:, 1], c=kmeans.labels_)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)

X_scaled
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4)

kmeans.fit(X_scaled)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=kmeans.labels_)
```

# Q2

```python
import pandas as pd

data = {
    'vehicle_serial_no': [5, 3, 8, 2, 4, 7, 6, 10, 1, 9],
    'mileage': [150000, 120000, 250000, 80000, 100000, 220000, 180000, 300000, 75000, 280000],
    'fuel_efficiency': [15, 18, 10, 22, 20, 12, 16, 8, 24, 9],
    'maintenance_cost': [5000, 4000, 7000, 2000, 3000, 6500, 5500, 8000, 1500, 7500],
    'vehicle_type': ['SUV', 'Sedan', 'Truck', 'Hatchback', 'Sedan', 'Truck', 'SUV', 'Truck', 'Hatchback', 'SUV']
}

# Create the DataFrame
df = pd.DataFrame(data)
```

```python
X = df.drop(columns=['vehicle_serial_no', 'vehicle_type'])

X
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=5)

kmeans.fit(X)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X.values[:, 0], X.values[:, 1], c=kmeans.labels_)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)

kmeans.fit(X_scaled)
```

```python
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=kmeans.labels_)
```

```python
import pandas as pd

df = pd.read_csv('student_performance.csv')
```

```python
df
```

```python
X = df[['weekly_self_study_hours', 'attendance_percentage', 'class_participation']]

X
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X_scaled[:, 0], X_scaled[:, 1], s =0.1)
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=5)

kmeans.fit(X_scaled)
```

```python
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], s=0.1, c=kmeans.labels_)
```
