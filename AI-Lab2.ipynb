```python
# --- TASK 1 ---

class SecurityAgent:
    def __init__(self, id, name, status):
        self.agent_id = id
        self.name = name
        self.status = status
        
class FirewallAgent(SecurityAgent):
    def __init__(self, id, name, status):
        super().__init__(id, name, status)

    def monitor_traffic(self):
        print("Firewall agent is monitoring traffic")

class MalwareDetectionAgent(SecurityAgent):
    def __init__(self, id, name, status):
        super().__init__(id, name, status)

    def scan_files(self):
        print("Malware detection agent is scanning files")

class AutomationAgent(SecurityAgent):
    def __init__(self, id, name, status):
        super().__init__(id, name, status)

    def run_automation(self):
        print("Automation agent is running automation")
        
    
```

```python
firewall = FirewallAgent(1, "firewall", "active")
malwareAgent = MalwareDetectionAgent(2, "malware", "active")
automation = AutomationAgent(3, "automation", "active")
```

```python
firewall.monitor_traffic()
```

```python
malwareAgent.scan_files()
```

```python
automation.run_automation()
```

```python
# --- TASK 2 ---

class Threat:
    def __init__(self, id, name, severity):
        self.threat_id = id
        self.name = name
        self.severity = severity
        
class PhishingThreat(Threat):
    def analyze_email(self):
        print("Emails are being analyzed")

class RansomwareThreat(Threat):
    def scan_files(self):
        print("Your files are being monitored")

class BotnetThreat(Threat):
    def detect_traffic(self):
        print("Traffic is being detected")

```

```python
phishing = PhishingThreat(1, "phishing", "high")
ransomware = RansomwareThreat(1, "ransomeware", "high")
botnet = BotnetThreat(1, "botnet", "low")
```

```python
phishing.analyze_email()
```

```python
ransomware.scan_files()
```

```python
botnet.detect_traffic()
```

```python
# --- TASK 3 ---

class Agent:
    def execute_response(self):
        pass
        
class AlertAgent(Agent):
    def execute_response(self):
        print("Alert Agent is sending notification")

class BlockAgent(Agent):
    def execute_response(self):
        print("Block Agent is blocking malicious activities")

class RecoverAgent(Agent):
    def execute_response(self):
        print("Recover agent is restoring affected systems")

alert = AlertAgent()
block = BlockAgent()
recover = RecoverAgent()

alert.execute_response()
block.execute_response()
recover.execute_response()
```

```python
# --- TASK 4 ---

class Employee:
    def work(self):
        pass

class Manager(Employee):
    def work(self):
        print("Manager is planning and managing the team")

class Developer(Employee):
    def work(self):
        print("Developer is writing and debugging code")

class Designer(Employee):
    def work(self):
        print("Designer is creating UI/UX designs")

employees = [ Manager(), Developer(), Designer()]

for employee in employees:
    employee.work()
```

```python
# --- Task 5 ---

class Student:
    def __init__(self, name):
        self.name = name
        self.__grade = None

    def set_grade(self, grade):
        if 0 <= grade <= 100:
            self.__grade = grade
        else:
            print("Invalid grade")

    def get_grade(self):
        return self.__grade

    def display_info(self):
        print("Student Name:", self.name, ", Grade:", self.__grade)


student1 = Student("Zaiyan")
student2 = Student("Umer")

student1.set_grade(95)
student2.set_grade(88)

print(student1.name, "'s grade:", student1.get_grade())
print(student2.name, "'s grade:", student2.get_grade())

student1.display_info()
student2.display_info()
```
