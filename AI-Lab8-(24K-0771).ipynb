## Q1

```python
import math

class Node:
    def __init__(self, value):
        self.value = value
        self.children = []
        self.minimax = None

class MinimaxAgent:
    def __init__(self):
        self.best_action = None

    def formulateGoal(self, node):
        return "Goal Reached" if node.minimax is not None else "Searching"
    
    def run_agent(self, node, env):
        status = self.formulateGoal(node)
        if status == "Goal Reached":
            return f"Final Decision: {self.best_action} with value {node.minimax}"
        else:
            result = env.minimax(node)
            return f"Robot chooses {self.best_action} (Minimax Value: {result})"

class Environment:
    def __init__(self, root, agent):
        self.root = root
        self.agent = agent

    def minimax(self, node, maxPlayer=True):
        if not node.children:
            return node.value
        
        if maxPlayer:
            maxEva = -math.inf
            print(f"Robot (MAX) evaluating options at {node.value}...")
            for n in node.children:
                eva = self.minimax(n, False)
                print(f"  > Decision {n.value} leads to expected safety: {eva}")
                if eva > maxEva:
                    maxEva = eva
                    if node == self.root:
                        self.agent.best_action = n.value
            node.minimax = maxEva
            return maxEva
        else:
            minEva = math.inf
            for n in node.children:
                eva = self.minimax(n, True)
                minEva = min(eva, minEva)
            node.minimax = minEva
            return minEva

za_opt1, za_opt2 = Node(3), Node(5)
zb_opt1, zb_opt2 = Node(2), Node(9)

zone_a = Node('Zone A')
zone_a.children = [za_opt1, za_opt2]

zone_b = Node('Zone B')
zone_b.children = [zb_opt1, zb_opt2]

robot_start = Node('Warehouse Entry')
robot_start.children = [zone_a, zone_b]

agent = MinimaxAgent()
env = Environment(robot_start, agent)
print(agent.run_agent(robot_start, env))
```

```python
# Q2 (b, c, d)
import math

class Node:
    def __init__(self, value):
        self.value = value
        self.children = []
        self.minimax = None
        self.alpha = -math.inf
        self.beta = math.inf
    
class AlphabetaAgent:
    def __init__(self):
        self.best_node = None

    def formulateGoal(self, node):
        return "Goal Reached" if node.minimax is not None else "Searching"
    
    def run_agent(self, node, env):
        status = self.formulateGoal(node)
        if status == "Goal Reached":
            return f"Goal Reached with value {node.minimax}"
        else:
            print(f"--- Starting Alpha-Beta Search at Root: {node.value} ---")
            return env.alphabeta(node)

class Environment:
    def __init__(self, root, agent):
        self.root = root
        self.agent = agent

    def alphabeta(self, node, alpha=-math.inf, beta=math.inf, maxPlayer=True):
        if not node.children:
            print(f"Terminal Node {node.value} reached.")
            return node.value
        
        if maxPlayer:
            maxEva = -math.inf
            print(f"MAX Node {node.value}: Initial alpha={alpha}, beta={beta}")
            for n in node.children:
                eva = self.alphabeta(n, alpha, beta, False)
                if eva > maxEva:
                    maxEva = eva
                    if node == self.root:
                        self.agent.best_node = n.value
                alpha = max(alpha, maxEva)
                if alpha >= beta:
                    print(f"!!! PRUNING at {node.value}: alpha({alpha}) >= beta({beta}). Skipping remaining children.")
                    break
            node.minimax = maxEva
            return maxEva
        else:
            minEva = math.inf
            print(f"MIN Node {node.value}: Initial alpha={alpha}, beta={beta}")
            for n in node.children:
                eva = self.alphabeta(n, alpha, beta, True)
                minEva = min(minEva, eva)
                beta = min(beta, minEva)
                if alpha >= beta:
                    print(f"!!! PRUNING at {node.value}: alpha({alpha}) >= beta({beta}). Skipping remaining children.")
                    break
            node.minimax = minEva
            return minEva

l6, l7 = Node(6), Node(7)
l2, l4 = Node(2), Node(4)
l8, l3 = Node(8), Node(3)

stop = Node('Stop'); stop.children = [l6, l7]
go = Node('Go'); go.children = [l2, l4]
turn = Node('Turn'); turn.children = [l8, l3]

start = Node('Start'); start.children = [stop, go, turn]

agent = AlphabetaAgent()
env = Environment(start, agent)
agent.run_agent(start, env)

print("\n--- Results ---")
print(f"Best Step for Car: {agent.best_node}")
print(f"Root Minimax Value: {start.minimax}")
```
