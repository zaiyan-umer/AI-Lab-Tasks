```python
# --- Task 1 ---
# part a
graph = {
    'A': ['B', 'C'],
    'B': ['D'],
    'C': ['F'],
    'D': ['H'],
    'F': ['I'],
    'H': ['J'],
    'I': [],
    'J': []
}

def dls(graph, node, goal, depth_limit, depth=0, path=[]):
    if depth > depth_limit:
        return None
        
    path.append(node)
    
    if node == goal:
        return path

    for n in graph[node]:
        if n not in path:
            if dls(graph, n, goal, depth_limit, depth+1, path) :
                return path
    path.pop()
    return None
```

```python
class Environment:
    def __init__(self, graph):
        self.graph = graph

class GoalBasedAgent:
    def __init__(self, goal):
        self.goal = goal

    def search(self, node, env, depth_limit):
        path = []
        path = dls(env.graph, node, self.goal, depth_limit, 0, path)
        if path:
            print("Goal found, path: ", path)
        else:
            print("Goal not found within depth limit")

env = Environment(graph)
agent = GoalBasedAgent('J')
agent.search('A', env, 5)
```

```python
# part b
class Graph:
    def __init__(self, graph):
        self.g = graph

    def ucs(self, src, dest):
        import heapq

        frontier = [(0, src, [src])]
        visited = set()

        while frontier:
            current_cost, current_node, path = heapq.heappop(frontier)

            if current_node in visited:
                continue
            visited.add(current_node)

            if current_node == dest:
                return path, current_cost

            for neighbor, cost in self.g.get(current_node, []):
                if neighbor not in visited:
                    heapq.heappush(frontier,
                                   (current_cost + cost,
                                    neighbor,
                                    path + [neighbor]))

        return None, float('inf')

class Environment:
    def __init__(self, graph):
        self.graph = graph

class UtilityBasedAgent:
    def __init__(self, goal):
        self.goal = goal

    def search(self, node, env):
        path, total_cost = env.ucs(node, self.goal)
        if path:
            print(f"Goal found! Path: {path}, Total cost: {total_cost}")
        else:
            print("Goal not reachable")

graph_data = {
    'A': [('B', 2), ('C', 5)],
    'B': [('D', 4)],
    'C': [('D', 1)],
    'D': []
}

env = Graph(graph_data)
agent = UtilityBasedAgent(goal='D')

agent.search('A', env)

```

```python
# --- Task 2 ---
from itertools import permutations

cities = ['1', '2', '3', '4']
graph = {
    '1': {'1':0, '2':10, '3':15, '4':20},
    '2': {'1':10,'2':0,  '3':35, '4':25},
    '3': {'1':15,'2':35, '3':0,  '4':30},
    '4': {'1':20,'2':25, '3':30, '4':0 }
}

def tsp(cities, graph):
    start = cities[0]
    min_cost = float('inf')
    best_path = None

    for perm in permutations(cities[1:]):
        path = [start] + list(perm) + [start]
        cost = sum(graph[path[i]][path[i+1]] for i in range(len(path)-1))
        if cost < min_cost:
            min_cost = cost
            best_path = path

    return best_path, min_cost

path, cost = tsp(cities, graph)
print("Shortest route:", path)
print("Total cost:", cost)
```

```python
# --- Task 3 ---
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F', 'G'],
    'D': ['H'],
    'E': [],
    'F': ['I'],
    'G': [],
    'H': [],
    'I': []
}

def dls(graph, node, goal, depth):
    if depth == 0 and node == goal:
        return [node]
    if depth > 0:
        for neighbor in graph.get(node, []):
            path = dls(graph, neighbor, goal, depth-1)
            if path:
                return [node] + path
    return None

def iterative_deepening_DFS(graph, start, goal, max_depth=50):
    for depth in range(max_depth):
        path = dls(graph, start, goal, depth)
        if path:
            return path
    return None

path = iterative_deepening_DFS(graph, 'A', 'I', max_depth=10)
print("Path found:", path)
```
