```python
print("---Task 4---")

def discount(purchase_amount):
    if purchase_amount < 1000:
        return purchase_amount - purchase_amount * 0.05
    elif purchase_amount < 5000:
        return purchase_amount - purchase_amount * 0.1
    else:
        return purchase_amount - purchase_amount * 0.15
    
amount = int(input("Enter the total price: "))
print("Net payable amount is: ", discount(amount))
```

```python
print("---Task 5---")

def count(string):
    vowels = consonents = 0
    vowels_array = ['a', 'e', 'i', 'o', 'u']

    for char in string:
        if char in vowels_array:
            vowels += 1
        else:
            consonents += 1

    return {
        "total" : len(string),
        "vowels" : vowels,
        "consonents" : consonents
    }

string = input("Enter a string: ")
print(count(string))
```

```python
print("---Task 6---")

def calcSubstring(string): 
    longest = ""
    current = ""

    for ch in string:
        if ch not in current:
            current += ch
        else:
            while ch in current:
                current = current[1:]
            current += ch

        if len(current) > len(longest):
            longest = current

    return longest


string = input("Enter a string: ")
print("Longest substring without repeating characters: ", calcSubstring(string))
```

```python
print("---Task 7---")

def repeatedNumbers(arr):
    repeating = []
    not_repeating = []
    for num in arr:
        if num in not_repeating:
            if num not in repeating:
                repeating.append(num)
        else:
            not_repeating.append(num)

    return repeating
        

l = list(map(int, input("Enter numbers: ").split()))
print("Repeated Numbers: ", repeatedNumbers(l))
```
